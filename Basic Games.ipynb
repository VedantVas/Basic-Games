```python
#Game like snakes and ladders but without snakes and ladders
import random as r 
score1= 0
score2= 0
while True:
    player1=input("Press Enter for player1 chance:")
    dice=r.randint(1,6)
    if(score1+dice<=20):
        score1+=dice
        print(f"Player1 got {dice}")
        print(f"Score of player1 is {score1}")
    else:
        continue
    if(score1==20):
        print("Player1 won..")
        break
    
    player2=input("Press Enter for player2 chance:")
    dice=r.randint(1,6)
    if(score2+dice<=20):
        score2+=dice
        print(f"Player2 got {dice}")
        print(f"Score of player2 is {score2}")
    else:
        continue
    if(score2==20):
        print("Player2 won...")
        break
```

```python
#Above game made using functions
import random as r


player_scores = [0, 0]
player_names = ["Player 1", "Player 2"]


def roll_die():
    return r.randint(1, 6)


def has_won(player_score):
    return player_score == 20


def play_game():
    turn = 0
    while True:
        current_player = turn % 2  
        input(f"{player_names[current_player]}'s turn. Press Enter to roll the die...")

        roll = roll_die()
        print(f"{player_names[current_player]} get a {roll} score")
        
        
        if player_scores[current_player] + roll <= 20:
            player_scores[current_player] += roll
            print(f"{player_names[current_player]} is now at {player_scores[current_player]} points")
        else:
            print(f"{player_names[current_player]} opps!.. cannot add {roll} as it would exceed 20 points.")
        
        
        if has_won(player_scores[current_player]):
            print(f"{player_names[current_player]} has won the game with exactly 20 points!..Hurray!...")
            break
        
       
        turn += 1


        
play_game()
```

```python
#Number guessing game 
counts=3
while(counts>=1):
    counts-=1
    dice=r.randint(1,6)
    guess=int(input("Enter the number between 1 to 6 "))
    
    if(dice==guess):
        print("You won..")
        break
    elif(counts>=1):
        print(f"You have remaining {counts} counts")
        print("Try again!")
    else:
        print("You lost !!!")
```
